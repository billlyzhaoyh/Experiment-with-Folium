```python
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
f = open("ES.txt", "r")
postcode=[]
latlong=[]
for x in f:
    lat_long=[]
    data=x.split()
    postcode.append(int(data[1]))
    lat_long.append(float(data[-3]))
    lat_long.append(float(data[-2]))
    latlong.append(lat_long)
f.close()
```

```python
dictionary = dict(zip(postcode, latlong))
```

```python
df=pd.read_excel('data.xlsx')
```

```python
latlong_list=[]
for pc in df.origin_postal_code.tolist():
    latlong_list.append(dictionary[pc])
```

```python
df['lat_long']=latlong_list
```

```python
map_hooray = folium.Map(location=[40.187, -4.0364],
                    zoom_start = 6) 

heat_data = df['lat_long'].values.tolist()

# Plot it on the map
HeatMap(heat_data).add_to(map_hooray)

for index, info in df.iterrows():
    folium.Marker(location=info[-1],popup='Number delivered: '+str(info[1]),icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_hooray)
# Display the map
map_hooray
```

```python
map_hooray.save('test1.html')
```

```python
def draw_map(data, location_start=[40.187, -4.0364], zoom_start=6, tiles='OpenStreetMap', color='red'): 
    markers = []
    fmtr = "function(num) {return L.Util.formatNum(num, 3) + '  ';};"
    new_map = folium.Map(location=location_start, zoom_start=zoom_start, tiles=tiles)
    marker = folium.Marker(df['lat_long'].values.tolist(), popup=df['Num_Delivered'], 
                                tooltip=df['Num_Delivered'],
                                icon=folium.Icon(color=color, icon='info-sign'))
    marker.add_to(new_map)
    folium.LayerControl(collapsed=False).add_to(new_map)
    MousePosition(position='topright', separator=' | ', prefix="Mouse:", lat_formatter=fmtr, lng_formatter=fmtr).add_to(new_map)
    Fullscreen(position='topleft', title='Full Screen', title_cancel='Exit Full Screen', force_separate_button=False).add_to(new_map)
    return new_map
```
